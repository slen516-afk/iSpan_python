### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
| |物件轉JSON並寫入文件|
| |物件轉JSON字串|
| |讀取JSON文件|
| |將JSON字串轉成Python物件|

```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['greem', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}
print(fruit_dict)
```

```python
#引入套件
import json
# json.dumps()物件轉json字串(dumps)
with open('fruit_data.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(fruit_dict))
```

```python
#物件轉json之後直接寫入檔案
#json.dump()可指定路徑檔案,它直接幫我寫入
with open('fruit_data.json', 'w', encoding='utf-8') as f:
    json.dump(fruit_dict, f, indent=2)
```

```python
# json.loads() 讀取字串轉物件
with open('fruit_data.json', 'r', encoding='utf-8') as f:
    data = f.read()
    data_dict = json.loads(data)

print(data_dict)
print(data_dict['apple'])
```

```python
x = '[1, 2, 3]'
print(type(x))

x_list = json.loads(x)
print(type(x_list))
print(x_list[0])
```

```python
#json.loads() 讀取檔案直接轉物件
with open('fruit_data.json', 'r', encoding='utf-8') as f:
    data_dict = json.load(f)

print(data_dict)
print(data_dict["apple"])
```
