##### 讀取檔案 read( )

```python
# 印出 50 個 "="
print("="*50)
"""
with 語句
通常用在 資料庫讀取、檔案讀取或寫入、多線程
需要釋放資源的情況
"""


# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案),as f 是把前面的東西存入f
with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f:
    doc = f.read()
    print(doc)


# 印出 50 個 "="
print("="*50)

```

```python
# 一行一行讀取
with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f:
    doc = f.readlines()
    print(doc)
```

```python
# 資料處理
# 去除所有的\n
with open(file="../Module2/2-1.py",mode="r",encoding="utf-8") as f:
    doc = f.readlines()
    print(doc)

while "\n" in doc:  #去掉 單獨的 斜線n \n
    doc.remove('\n')
print(doc)

new_list = []  #去除 list裡的 斜線n \n
for doc in doc:
    new_list.append(doc.strip("\n"))

print(new_list)
```
