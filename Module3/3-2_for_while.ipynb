| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
#Ctrl + D 選擇相同的文字做修改
# for <變數> in <可迭代的物件>:
n = 1
for char in 'contents':
    print(char, end = "")
    print(n,end=',')
    n += 1
```

```python
# range(m)
for i in range(5):
    print(i)
#走訪0~4,python由0起算
```

```python
# range(n, m-1) 走訪n到m-1的數字
for i in range(1, 11):
    print(i, end = " ")

```

```python
# range(n, m-1, step)
for i in range(1, 11, 3):
    print(i)
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = 4
n = 5

for i in range(n):
    if a % 2 == 0:
        print(a+i*2)
    else:
        print(a+1+i*2)
```

```python
a = 4
n = 5
if a % 2 != 0:
    a += 1

for i in range(n):
    print(a+i*2)
```

```python
a = 5
n = 3

for i in range(n):
    print(a+1+i*2)
```

```python
n = 5
for i in range(n):                  # i = 0..n-1（第幾行）
    for s in range(i):              # 先印左側空白
        print(' ', end='')
    for k in range(n - i):          # 再印星號
        print('* ', end='')
    print()                         # 換行
```

```python
n = 8  # 行數
for i in range(1, n + 1):
    for s in range(i): 
        print(' ', end='')
    for k in range(n - i):
        print('* ', end = '')
    print()
```

```python
n = 5
for i in range(n, 1, -1):
    for s in range(i): #印出左邊空格
        print(' ',end='')
    for k in range(n-i):
        print('* ',end = '')
    print()
```

```python
a = 5
for i in range(a, 0, -1):
    for o in range(i):
        print(" " , end ="")
    for k in range(a-i):
        print("* ",end = "")
    print()
```

```python
a = 5
for i in range(a, 0, -1):
    for c in range(i):
        print(" ",end = "")
    for b in range(a-i):
        print("* ",end = "")
    print()
            
```

```python
n = 10
for a in range(n, 0, -1):
    for b in range(a):
        print(" ",end="")
    for c in range(a-b):
        print("* ",end="")
    print()
```

##### while

```python
count = 1

while count <= 5:
    print(count, end = " ")
    count += 1

print("count 已經不小於5 結束迴圈")

```

```python
count = 0
while count<=5: #一定要給while結束條件
    print(count, end=" ")
    count +=1

print(count)
```
