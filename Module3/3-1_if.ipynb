條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
num = 20
if num > 10:
    print("num > 10")
```

##### if...else...

```python
num = 20
if num > 10:
    print("num > 10")
else:
    print("num 不大於 10")
```

##### if...elif... (else)

```python
num = 10.5
print(type(num))
```

```python
num = 6
if num > 10:
    print("num > 10")
elif num % 2 == 0:
    print("num is even")
elif type(num) == int:
    print("num 資料型態為 int")

else:
    print("num 不大於 10 且不為 int")
```
