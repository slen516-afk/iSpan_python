##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
for i in "content":
    if i == "t":
        break
    print(i)
print("迴圈結束")
```

```python
for i in range(1, 101):    
    print(i)
```

```python
for char in "content":
    if char == "t":
        continue
    print(char)

print("迴圈結束")
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
for char in "content":
    if char == "t":
        continue
    print(char, end = "")
```

### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行

```python
A = 0
for amount in range(1, 600):
    if A >= 500:
        break
    A += amount
print(A)
```

```python
cnt = 0
for i in range(1, 101):
    print(i, end=" ")
    cnt += i
    if cnt > 500:
        break

print() #換行
print("="*50)
print(cnt)
print("="*50)
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
A = 0
for amount in range(1, 101, 2):
    if amount % 5 == 0:
        continue
    A += amount
    
print(A)
```

```python
sum = 0
for i in range(1, 101):
    if i % 2 == 0 or i % 5 == 0:
        continue
    print(i, end=", ")
    sum += i
print()
print(sum)
```
