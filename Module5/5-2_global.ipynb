```python
# 全域變數
name = 'Ryan'

# 定義函式
def greeting():
    name = 'user'
    msg = 'hello'

    return msg + ' ' + name

# 執行函式
data = greeting()
print(data)

# 輸出 name

```

##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = 'Ryan'

# 定義函式
def greeting():
    # name = 'user'
    global name #呼叫全域變數
    msg = 'hello'

    return msg + ' ' + name

# 執行函式

data = greeting()
print(data)
# 輸出 name
print(name)

```
