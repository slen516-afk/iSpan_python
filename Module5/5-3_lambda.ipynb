### lambda 匿名函式  
用法  

`function_name = lambda arguments: expression`

````python
# 建立函式
def add_num(x, y):
    return x + y

print(add_num(6, 10))
````

````python
# 基本用法
add_num = lambda x, y: x + y
print(add_num(6, 19))
````

````python
# 條件判斷
def check_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
    
x= check_even(8)
print(x)
print(check_even(5))
````

````python
#lambda = 簡短版的function(def)
check_even = lambda num:True if  num % 2 == 0 else False
print(check_even(4))
````

````python
outside = []
for i in range(1, 11):
    if i % 3 == 0:
        outside.append([1, 5, i])
    else:
        outside.append([1, 2, i])

print(outside)
outside.sort(reverse=True) #把([1, 2 ,i])跟([1, 5, i])分開來排在一起做整理
print(outside)#印出第二列
outside.sort(key = lambda a : a[2], reverse=True)#[2]=index 2 #lambda接參數:表達式,#lambda在此的功能是以list的index2的大到小做排序整理
print(outside)
````
