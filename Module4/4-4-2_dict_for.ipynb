```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
color_data = {} #空dictionary
color_data["apple"] = "red"
color_data["apple"] = "green"
print(color_data)
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
color_data = {}
for fruit_color in fruit_color_list:
    print(fruit_color)

    for name, color in fruit_color.items():
        #如果沒有該key值就執行...
        if color_data.get(name) == None:
            color_data[name] = [color]
        #有該key值就執行
        else:
            color_data[name].append(color)

print(f"{color_data}")
```

##### `defaultdict(<type>)`

```python
from collections import defaultdict
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
data = defaultdict(int)

# 以int型態的預設值 0 來初始化
print(data["price"])
print(data)
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}
    ]
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
from collections import defaultdict
color_data = defaultdict(list)
color_data["apple"].append("red")
print(color_data)

for fruit_color in fruit_color_list:
    print(fruit_color)
    for name, color in fruit_color.items():
        color_data[name].append(color)
print(color_data)

# for fruit_color in fruit_color_list:
#     for name, color in fruit_color.items():
#         color_data[name].append(color)
# print(color_data)
```
