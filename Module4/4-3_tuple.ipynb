```python
# tuple 初始化: 第一種
data = ("Big", "Data", "is", "Good")
print(data)
print(type(data))
```

```python
# tuple 初始化: 第二種
data = "Jarvis", 18
print(data)
print(type(data))

# name, age = "Jarvis", 18
# print(name)
# print(age)
```

```python
data = "Jarvis", 18, "Taipei"
print(data)

name, age, _class = data
print(name)
print(age)
print(_class)
```

```python

# 透過指定索引輸出值
data = "Jarvis", 18, "Taipei"
print(data)
print(data[2])

```

```python
# tuple 不可以修改指定索引的值
data[0] = 100
```

```python
# (list)可以修改指定索引的值
num_list = [1, 2, 3]
num_list[0] = 100
print(num_list)
```

```python
# 用 for 迴圈逐一輸出資料
data = "Jarvis", 18, "AIPE02"
for i in data:
    print(i)
```

```python
# 用 len 計算 tuple 資料個數
data = "Jarvis", 18, "AIPE02"
print(len(data))
print(data[2])
```
