```python
names = ["John", "Mary", "Bob"]
scores = [80, 95, 70]

score_dict = {}

for i in range(len(names)): #得到[0],[1],[2],而不是"John","Mary","bob"跟80 95 70
    name = names[i] #因為i會走訪[1],[2],[3],所以會是"John", "Mary", "Bob
    score = scores[i] #會是80,95,70
    score_dict[name] = score #把dic塞進空set的方式
score_dict
```

```python
names   = ["John", "Mary", "Bob"]
chinese = [80, 90, 70]
english = [75, 95, 85]
math    = [88, 92, 78]
score_dict = {}
for i in range(len(names)):
    name = names[i]
    chn = chinese[i]
    eng = english[i]
    ma = math[i]

    score_dict[name] = {
        'chinese':chn,
        'english':eng,
        'math':ma
    }
    

print(score_dict)
```
