##### slice 透過 [ : ] 進行分割
`[n:m]` 走訪 n 到 m - 1 的元素

```python
# 初始化一個字串
data = "Hello,World!"
print(data)
print(data[0])
print(data[6])
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data[4])
print(data[0:5])
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data[4])
print(data[:5])
```

```python
# 從 6 取到尾
print(data[6:])
```

```python
# 從 7 取到 9（不包含第 9 個元素，實際為 7 ~ 8）
print(data[7])
print(data[7:9])
```

##### 使用負號

```python
# 從倒數第 10 取到倒數第 7（不包含倒數第 7 的元素）
data = "Hello,World!"
print(data[-10:-7])
```

```python
# 從倒數第 6 取到尾

data = "Hello,World!"
print(data[-6:])
```

```python
# 從頭取到倒數第 7（不包含倒數第 7 的元素）
data = "Hello,World!"
print(data[:-7])
```

```python
# 也可以使用在 list
num_list = [i for i in range(1, 11)]
print(num_list[-1]) #看最後一筆資料
print(num_list[-3:])#看末三筆資料
print(num_list[:3]) #看前三筆資料
```
