|語法|說明|
|:---:|:---:|
|%s |插入 字串 的值到字串中|
|%d |插入 數字 的值到字串中|
|%f|插入 浮點數 的值到字串中|
| %% |輸出 % (類似跳脫字元)|

```python
# 多組文字
data1 = {"name": "Jarvis"}
data2 = [1,2,3]
msg = "%s, %s!" %("Hello","World")
msg = "Message: %s, %s!" %(data1, data2)
print(msg)
```

```python
# 整數

age = 18
msg = "I am %d years old. " %(age)
msg = '資料輸出的結果 Age = %d ' %(age)
print(msg)

print(age)
```

```python
# 文字與整數
msg = "%s is %d years old." %("Jarvius",18)
print(msg)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % "Hello"
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)

```

```python
# 指定浮點數位數
msg = '[%.5f]' % 3.14256
print(msg)

```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = '[%.2s]' %"Hello"
print(msg)

```

```python
# 空白補 0
msg = '[%0.10f]' % 3.14159
print(msg)

```
