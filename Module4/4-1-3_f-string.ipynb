```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "Jarvis"
age = 19
msg=f"{name} is {age} years old"
print(msg)
```

```python
now = "2025-11-01 15:52"
status = "INFO"
log = f"{now}[{status}] - {msg}"
print(log)
```

```python
# 靠左對齊、靠右對齊
name = "Jarvis"
age = 19

msg = f"name: [{name:<10}], age: [{age:>10}]"

print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
name = "Jarvis"
age = 19

msg = f"name: [{name:ㄏ<10}], age: [{age:ㄏ>10}]"

print(msg)
```

```python
# 置中對齊
name = "Jarvis"
age = 19

msg = f"name: [{name:^10}], age: [{age:^10}]"

print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
name = "Jarvis"
age = 19

msg = f"name: [{name:.3}]"

print(msg)

```

```python
data = 3.14159
print(f'{data:.2f}')
```
