```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_price_dict = {
    "apple": {
        'price':50,
        'quantity':20,
        'color':["red",
                 'green'
        ]
            
        },
    
    'orange':{
        'price':30,
        'quantity':30,
        'color': ['green',
                  'orange'
                ],

        },
    'tomato':{
        'price':60,
        'quantity':90,
        'color':["green",
                 "red"
                 ]
    }
}


        
int(fruit_price_dict['orange']['price'])


```

##### dict 的合併

```python
# dict 的合併
# dict.update()*
fruit_dict2 = {
    "kiwi":{
        "price": 30,
        "amount": 20,
        "color": ["green"]
    }
}
fruit_dict2.update(fruit_price_dict)
print(fruit_dict2)
```

```python
print(fruit_dict2)
fruit_dict3 = {
    "kiwi":{
        "price": 30,
        "amount": 20,
        "color": ["green"]
    **fruit_price_dict,
    },
     
    
}
print(fruit_dict3)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
# 打印出來的資料沒有按照順序
from pprint import pprint
pprint(fruit_dict3)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

pprint(fruit_dict3)
fruit_dict3['apple']['color'].append('yellow')
fruit_dict3['tomato']['color'].append('green')
pprint(fruit_dict3)
```
